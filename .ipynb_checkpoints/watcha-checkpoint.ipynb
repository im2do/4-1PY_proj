```python
import time
import re
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from konlpy.tag import Okt

class WatchaDataMiner:
    def __init__(self):
        self.okt = Okt()
        
        # 1. 제거할 단어(불용어) 사전 정의
        # 형태소 분석 시 기본형(어간)으로 변환되므로, '하다', '되다' 등의 원형을 등록합니다.
        self.stopwords = {
            # 시스템 및 기초 불용어
            '하다', '되다', '이다', '있다', '없다', '않다', '그렇다', '아니다', 
            '같다', '어떻다', '이렇다', '저렇다', '진짜', '너무', '정말', '그냥', 
            '이거', '저거', '그거', '많다', '좋다', '나오다', '보다', '가다',
            # 사용자 지정 영화/도메인 관련 불용어
            '케데헌', '골든', '케이팝데몬헌터스', '영화', '드라마', '케이팝',
            '작품', '관람', '평점', '리뷰', '감독', '배우', '관객'
        }

    def clean_and_tokenize(self, text):
        """텍스트 정제 및 형태소 분석 (명사, 형용사만 추출)"""
        # 1차 정제: 한글과 띄어쓰기만 남기고 특수기호, 이모지 등 모두 제거
        text = re.sub(r'[^가-힣\s]', ' ', text)
        text = re.sub(r'\s+', ' ', text).strip()
        
        if not text:
            return ""

        # 2차 정제: 형태소 분석 (stem=True로 어간 추출, ex: '하는' -> '하다')
        pos_result = self.okt.pos(text, stem=True)
        
        valid_words = []
        for word, pos in pos_result:
            # 명사(Noun)와 형용사(Adjective)만 추출, 동사(Verb)는 제외
            if pos in ['Noun', 'Adjective']:
                # 1글자 단어('아', '오', '음' 등) 및 불용어 사전에 있는 단어 제외
                if len(word) > 1 and word not in self.stopwords:
                    valid_words.append(word)
                    
        return " ".join(valid_words)

def scrape_watcha_massive(target_url, target_count=10000):
    """왓챠피디아 실시간 스크롤 및 데이터 수집 (DOM 가상화 대응)"""
    
    # 디버깅 모드로 열려있는 Mac 크롬과 연결
    chrome_options = Options()
    chrome_options.add_experimental_option("debuggerAddress", "127.0.0.1:9222")
    driver = webdriver.Chrome(options=chrome_options)
    
    miner = WatchaDataMiner()
    
    # 중복 수집 방지를 위해 Set 자료구조 사용
    raw_comments_set = set()
    processed_data = []
    
    print(f"[{target_url}] 데이터 수집을 시작합니다. 목표: {target_count}개")
    driver.get(target_url)
    time.sleep(5) # 페이지 최초 로딩 대기
    
    scroll_attempts = 0
    max_scroll_attempts = target_count // 5 # 무한 루프 방지용 최대 스크롤 횟수
    last_height = driver.execute_script("return document.body.scrollHeight")
    
    while len(raw_comments_set) < target_count and scroll_attempts < max_scroll_attempts:
        # 화면에 보이는 코멘트 요소들 추출 (왓챠 클래스명: F12로 확인 후 필요시 업데이트)
        elements = driver.find_elements(By.CSS_SELECTOR, "div.css-1f9p2v7 span.css-1i0u70")
        
        current_batch_count = 0
        for elem in elements:
            try:
                raw_text = elem.text.strip()
                # 새로운 코멘트일 경우에만 처리
                if raw_text and raw_text not in raw_comments_set:
                    raw_comments_set.add(raw_text)
                    
                    # 수집과 동시에 정제 작업 수행
                    cleaned_text = miner.clean_and_tokenize(raw_text)
                    if cleaned_text: # 정제 후 남은 단어가 있는 경우만 추가
                        processed_data.append({
                            'raw_text': raw_text,
                            'cleaned_words': cleaned_text
                        })
                    current_batch_count += 1
            except Exception:
                continue
                
        # 스크롤 내리기
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2.5) # 왓챠 서버 부하 및 렌더링 대기 (너무 짧으면 차단 위험)
        
        # 진행 상황 출력
        scroll_attempts += 1
        if scroll_attempts % 10 == 0:
            print(f"> 스크롤 {scroll_attempts}회 진행 | 현재까지 수집된 고유 코멘트: {len(raw_comments_set)}개")
            
        # 더 이상 스크롤이 내려가지 않는지 체크 (페이지 끝 도달)
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            print("페이지의 끝에 도달했거나 더 이상 로드할 데이터가 없습니다.")
            break
        last_height = new_height

    print(f"\n데이터 수집 및 정제 완료! (총 {len(processed_data)}건의 유효 데이터 추출)")
    
    # DataFrame 변환 및 저장
    df = pd.DataFrame(processed_data)
    return df

# ==========================================
# 실행 파트
# ==========================================
if __name__ == "__main__":
    # 왓챠피디아 코멘트 URL 입력 (반드시 '/comments' 로 끝나는 코멘트 전체 보기 페이지여야 함)
    WATCHA_URL = "https://pedia.watcha.com/ko/contents/mW42oMq/comments?order=recommended"
    
    # 데이터 수집 실행 (목표: 10,000개)
    result_df = scrape_watcha_massive(WATCHA_URL, target_count=10000)
    
    # CSV로 저장 (Mac/Windows 모두 한글 깨짐 방지를 위해 utf-8-sig 사용)
    file_name = "kpop_demon_hunters_watcha_cleaned.csv"
    result_df.to_csv(file_name, index=False, encoding='utf-8-sig')
    print(f"[{file_name}] 파일이 성공적으로 저장되었습니다.")
    
    # 상위 5개 결과 확인
    print(result_df.head())
```
